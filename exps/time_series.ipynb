```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
plt.rcParams.update({'font.size': 14})
```

# Load data

```python
df = pd.read_csv(
    '/u/yangkc/natesilver538_botscore.csv',
    sep='\t',
    names=['date', 'bot_score'],
    parse_dates=['date']
)
```

```python
df['date'] = df.date.apply(lambda x: x.replace(tzinfo=None))
```

```python
df.sort_values(by='date', inplace=True)
```

```python
df['day_since'] = (df.date - pd.to_datetime('2020-09-01')).dt.days
```
